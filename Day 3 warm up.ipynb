## Word Counts using Dictionary

```python
words_list = ['Data', 'AI', 'Artificial', 'AI', 'AI', 'ML']
word_counts = {}
for word in words_list:
  if word in word_counts:
     word_counts[word] +=1
  else:
     word_counts[word] = 1
print(word_counts)
```

## Grouping Items by Category Using a Dictionary

```python
items = [('French', 'bonjour'), ('Portuguese', 'obrigado'), ('English', 'hello'), ('English', 'rendezvous'), ('English', 'Ok')]
grouped_items= {}
for cat, item in items:
  if cat not in grouped_items:
     grouped_items[cat]=[]
  grouped_items[cat].append(item)
  #else:
    #pass # your code
print(grouped_items)
```

## Calculating Average using Dictionary

```python
data = {'Alic': [80, 90, 100], 'Bob': [50, 80, 99]}
avg_score = {}
for key, values in data.items():
  avg_score[key] = round(sum(values)/len(values),2)

print(avg_score)
```
